```python
import sklearn 
```

```python
from sklearn import datasets 
```

```python
dir(datasets) # to see which datasets are there 
```

```python
data=datasets.load_boston()
```

```python
dir(data)
```

```python
data.feature_names
```

```python
X=(data.data)
```

```python
y=data.target
```

```python
from sklearn import (linear_model,svm)
```

```python
dir(linear_model)
```

```python
model=linear_model.LinearRegression()
```

```python
from sklearn import feature_selection
dir(feature_selection)
```

```python
help(feature_selection.SelectFromModel)
```

```python
from sklearn.feature_selection import SelectKBest, chi2,SelectFromModel
from sklearn.linear_model import Lasso
#X_new= SelectKBest(chi2,k=7).fit_transform(data.data, data.target)
```

```python
#help(Lasso)
```

```python
selector = SelectFromModel(estimator=Lasso(alpha=0.5)).fit(X, y)
```

```python
selector.estimator_.coef_
```

```python
X_new=selector.transform(X)
```

```python
X_new.shape
```

```python
model #default parameters set displayed 
```

```python
from sklearn.model_selection import train_test_split
```

```python
# X_new is the transformed dataset 
X_train, X_test, y_train, y_test = train_test_split(X_new, data.target, test_size=0.25, random_state=0)
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
